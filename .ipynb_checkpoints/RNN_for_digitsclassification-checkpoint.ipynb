```python
import torch
import torch.nn as nn
import torch.optim as optim 
import torchvision
from torchvision.transforms import transforms
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean = [0.5], std = [0.5])
])
```

```python
train_set = MNIST("./dataa", train = True, transform = transform)
test_set = MNIST("./dataa", train = False, transform = transform)
```

```python
train_set
```

```python
trainloader = DataLoader(train_set, batch_size = 64, shuffle = True) 
testloader = DataLoader(test_set, batch_size = 64) 
```

```python
trainloader
```

```python
# Image size grayscale = 1x28x28 
```

```python
input_size = 28
hidden_size = 128
num_layers = 2
out_features = 10
```

```python
class RNN(nn.Module):

    def __init__(self):
        super().__init__()
        
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.rnn = nn.RNN(
            input_size = input_size, 
            hidden_size = hidden_size,
            num_layers = num_layers,
            batch_first = True 
        )

        self.fc = nn.Linear(hidden_size, out_features = 10) 

    def forward(self, x):
        x = x.squeeze(1) 
        
        # optinal = > shape (num_of_layers, batch_size, hidden_size) 
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        output, _ = self.rnn(x, h0)

        output = self.fc(output[:, -1, :])

        return output 

        
```

```python
model = RNN()

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters()) 
```

```python
epochs = 10

for epoch in range(epochs):
    model.train()
    epochs_training_loss = 0.0

    for images, labels, in trainloader:
        optimizer.zero_grad()

        output = model(images)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()

        epochs_training_loss += loss.item() 

    print(f" epoch {epoch+1}/{epochs} & loss {epochs_training_loss/len(trainloader)}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix

model.eval()

y_true = []
y_pred = []

with torch.no_grad():
    for images, labels, in testloader:
        outputs = model(images) 

        _, predicted = torch.max(outputs, 1)

        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

acc = accuracy_score(y_true, y_pred)
print("Accuracy:- ", acc*100)

pre = precision_score(y_true, y_pred, average = 'weighted')
print("Precision:- ", pre*100) 

conf_matrix = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:- ", conf_matrix*100) 
```
